# Пример расчета и визуализации с использованием сложной формулы


В этом примере мы предположим, что одно (или несколько) из измерений проводятся не напрямую, а косвенно. В таком случае необходимо сначала провести расчет значения по формуле, а потом построить искомую зависисмость. В качестве примера мы выполним небольшое моделирование, в котором весь массив данных будет получаться расчетным способом, что может быть применимо для сравления результатов с теоретическими значениями.

### Задача
Рассмотрим закон охлаждения Ньютона:

$$ T(t) = T_\text{env} + (T_0 - T_\text{env}) e^{-kt} $$

где:
- $T(t)$ — температура объекта в момент времени $t$,
- $T_\text{env}$ — температура окружающей среды,
- $T_0$ — начальная температура объекта,
- $k$ — коэффициент теплопередачи,
- $t$ — время.

Мы будем рассчитывать температуру объекта $T(t)$ в зависимости от времени $t$ и визуализировать результаты.


## 1. Расчет температуры объекта

```python

import numpy as np

# Данные
T_env = 25  # Температура окружающей среды (°C)
T_0 = 100   # Начальная температура объекта (°C)
k = 0.1     # Коэффициент теплопередачи (1/с)
time = np.linspace(0, 100, 500)  # Время (с)

# Расчет температуры объекта
T = T_env + (T_0 - T_env) * np.exp(-k * time)

# Вывод первых нескольких значений
for t, temp in zip(time[:10], T[:10]):
    print(f"Время: {t:.1f} с, Температура: {temp:.2f} °C")
```

## 2. Построение графика зависимости температуры от времени

```python

import matplotlib.pyplot as plt

# Построение графика
plt.plot(time, T, color='red', label='Закон охлаждения Ньютона')
plt.xlabel('Время (с)')
plt.ylabel('Температура объекта $T(t)$ (°C)')
plt.title('Зависимость температуры объекта от времени')
plt.grid(True)
plt.legend()
plt.show()
```

## 3. Влияние коэффициента теплопередачи

Рассмотрим также дял примера, как влияет коэффициент теплопередачи на результат. Для этого построим аналогичные зависимости для нескольких коэффициентов теплопередачи.

```python

# Коэффициенты теплопередачи
k_values = [0.05, 0.1, 0.2]

plt.figure(figsize=(8, 6))
for k in k_values:
    T = T_env + (T_0 - T_env) * np.exp(-k * time)
    plt.plot(time, T, label=f'k = {k} (1/с)')

plt.xlabel('Время (с)')
plt.ylabel('Температура объекта $T(t)$ (°C)')
plt.title('Влияние коэффициента теплопередачи на охлаждение')
plt.grid(True)
plt.legend()
plt.show()
```


### Итог
Мы рассчитали температуру объекта на основе закона охлаждения Ньютона, визуализировали её зависимость от времени и 
исследовали влияние коэффициента теплопередачи.
